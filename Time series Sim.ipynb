```r
w1=rnorm(1000,0,1)
w2=rexp(1000,1)
```

```r
sigma1=sqrt(var(w1))
sigma2=sqrt(var(w2))

ws1=(w1-mean(w1))/sigma1
ws2=(w2-mean(w2))/sigma2


w1 = ts(ws1,start=1)
w2 = ts(ws2,start=1)




par(mfrow=c(2,2))
ts.plot(w1)
ts.plot(w2)

acf(w1, lag.max=10)
acf(w2, lag.max=10)
```

```r
w1=rnorm(502)
w2=rexp(502)-1

a1=c(1,-0.5,0.2)
a2=c(1,0.5,0.2)
```

```r
ma2.11=filter(w1, filter=a1,side=1)
ma2.11=ma2.11[3:502]


ma2.12=filter(w1, filter=a2,side=1)
ma2.12=ma2.12[3:502]


ma2.21=filter(w2, filter=a1,side=1)
ma2.21=ma2.21[3:502]

ma2.22=filter(w1, filter=a2,side=1)
ma2.22=ma2.22[3:502]

par(mfrow=c(2,2))
acf(ma2.11)
acf(ma2.12)
acf(ma2.21)
acf(ma2.22)



pacf(ma2.11)
pacf(ma2.12)
pacf(ma2.21)
pacf(ma2.22)
```

```r
w1=rnorm(502)
a4=c(1,-0.2,0.8,1.2)

ma2.4=filter(w1*(2*(1:502)+0.5), filter=a4, side=1)

ma2.4=ma2.4[4:502]

par(mfrow=c(2,2))

ma24.s=ts(ma2.4, start=1)
plot(ma24.s)

acf(ma24.s)
#pacf(ma24.s, lag.max=10)
```
