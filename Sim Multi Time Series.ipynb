```r
e=rnorm(501)
x=rnorm(502)
xmat=cbind(x[2:501],x[1:500])
ys=0

for(k in 1:500){
    if (k==1){
        
        ys[k]=e[k]+0.6*xmat[k,1]+0.3*xmat[k,2]}
    
    if (k>1){
        
        ys[k]=0.8*ys[k-1]+e[k]+0.6*xmat[k,1]+0.3*xmat[k,2]}
}

ys=ts(ys[251:500])
xs=ts(x[253:502])
acf(ts.union(ys,xs))
```
